```python
import pandas as pd
import numpy as np
import string
from textblob import TextBlob
import emoji
from langdetect import detect, LangDetectException
import re
```

```python
new_df_restorant_commit=pd.read_csv(r"C:\Software_project\Restorant_KDS\Comment_Scraping\restorant_pizza_doner_yorumları_cleared.csv",encoding='utf-8')
print(new_df_restorant_commit.head())
```

```python
new_df_restorant_commit = new_df_restorant_commit.apply(lambda x: x.str.lower())
print(new_df_restorant_commit.head())
```

```python
# new_df_restorant_commit["new_Yorum"]= new_df_restorant_commit["Yorum"]
# new_df_restorant_commit["new_Yorum"] = new_df_restorant_commit.to_csv("restorant_pizza_doner_yorumları.csv", index=False)
print(new_df_restorant_commit.columns)
print(new_df_restorant_commit.head())
```

```python
new_df_restorant_commit["new_Yorum"] = new_df_restorant_commit["new_Yorum"].apply(lambda x: emoji.replace_emoji(str(x), replace=''))
```

```python
# table1 = str.maketrans('', '', '0123456789')
# table2 = str.maketrans('', '', string.punctuation)
# table3=str.maketrans("","","!@#$%^&*()_+-=[]{};:\'\"\\|,<.>/?`~")

# new_df_restorant_commit["new_Yorum"] = new_df_restorant_commit["new_Yorum"].apply(
#     lambda x: str(x).translate(table1).translate(table2).translate(table3))
# new_df_restorant_commit["new_Yorum"]=new_df_restorant_commit["new_Yorum"].str.strip()
print(new_df_restorant_commit["new_Yorum"].head(5))
```

```python
new_df_restorant_commit.drop_duplicates(subset=['new_Yorum'], inplace=True)
new_df_restorant_commit = new_df_restorant_commit[
    new_df_restorant_commit['new_Yorum'].apply(lambda x: len(str(x).split()) >= 3)
]
print(new_df_restorant_commit['new_Yorum'].head(10))
new_df_restorant_commit.to_csv(r"C:\Software_project\Restorant_KDS\Comment_Scraping\restorant_pizza_doner_yorumları_cleared.csv",index=False,encoding="utf-8-sig")
```

```python


def is_turkish(text):
    try:
        dil = detect(text)
        # Eğer dil Türkçe ('tr') ise True döndür, değilse False
        return dil == 'tr'
    except LangDetectException:#metin kısaysa hata verir
        return False

new_df_restorant_commit = new_df_restorant_commit[
    new_df_restorant_commit['new_Yorum'].apply(is_turkish)
]

print(f"Türkçe olmayanlar silindi. Kalan satır sayısı: {len(new_df_restorant_commit)}")
```
