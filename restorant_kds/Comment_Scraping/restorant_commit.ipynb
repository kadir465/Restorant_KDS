```python
def fix_puan(x):
    # Eğer pandas Timestamp ise → bu tarih aslında ay.gün puanı
    if isinstance(x, pd.Timestamp):
        return float(f"{x.month}.{x.day}")
    
    # Eğer string olarak tarih geldiyse
    if isinstance(x, str):
        try:
            ts = pd.to_datetime(x)
            return float(f"{ts.month}.{ts.day}")
        except:
            pass
    
    # Eğer zaten sayı ise → direkt döndür
    return x

df_restornt_bilgi["Puan"] = df_restornt_bilgi["Puan"].apply(fix_puan)

print(df_restornt_bilgi["Puan"].head())
df_restornt_bilgi=df_restornt_bilgi.to_csv("restorant_bilgileri.csv",index=False,encoding="utf-8-sig")
```
