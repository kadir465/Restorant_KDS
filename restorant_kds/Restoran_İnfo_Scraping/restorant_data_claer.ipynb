```python
import pandas as pd
import numpy as np
```

```python
new_df_restorant_bilgi=pd.read_csv(r"restorant_bilgileri.csv")
print(new_df_restorant_bilgi.head())
```

```python
new_df_restorant_bilgi["min siparis"]=new_df_restorant_bilgi[["Min. Tutar",'Minimum Tutar','Minimum Sipariş']].astype(str).agg(" ".join,axis=1)
print(new_df_restorant_bilgi["min siparis"].head())

col=new_df_restorant_bilgi["min siparis"].astype(str)
col=col.str.replace("nan","",regex=False).str.strip()
new_df_restorant_bilgi["min siparis"]=col
print(new_df_restorant_bilgi["min siparis"].head())

new_df_restorant_bilgi["min siparis"]=new_df_restorant_bilgi["min siparis"].str.replace("TL","",regex=False).str.strip()
print(new_df_restorant_bilgi["min siparis"].head())
```

```python
new_df_restorant_bilgi=new_df_restorant_bilgi.drop(columns=["Min. Tutar",'Minimum Tutar','Minimum Sipariş'])
```

```python
print(new_df_restorant_bilgi.columns.to_list())
```

```python
new_df_restorant_bilgi["mutfak"]=new_df_restorant_bilgi[["Mutfak","Mutfak Tipi"]].astype(str).agg(" ".join,axis=1)
print(new_df_restorant_bilgi["mutfak"].head())

col=new_df_restorant_bilgi["mutfak"].astype(str)
col=col.str.replace("nan","",regex=False).str.strip()
new_df_restorant_bilgi["mutfak"]=col
print(new_df_restorant_bilgi["mutfak"].head())
```

```python
new_df_restorant_bilgi=new_df_restorant_bilgi.drop(columns=['Mutfak','Mutfak Tipi'])
new_df_restorant_bilgi=new_df_restorant_bilgi.to_csv("restorant_bilgileri.csv",index=False,encoding="utf-8-sig")
```

```python
new_df_restorant_bilgi["Puan"]=(
    new_df_restorant_bilgi["Puan"].astype(str)
    .str.replace("Belirtilmemiş","",regex=False)
    .str.replace(",","",regex=False).str.strip()
)
new_df_restorant_bilgi=new_df_restorant_bilgi.to_csv("restorant_bilgileri.csv",index=False,encoding="utf-8-sig")
```

```python
print(new_df_restorant_bilgi.columns.to_list())
puan_mean=new_df_restorant_bilgi["Puan"].mean()
puan_mean=round(puan_mean,2)
print(puan_mean)
```

```python
print(new_df_restorant_bilgi["Puan"].isnull().sum())
new_df_restorant_bilgi["Puan"]=new_df_restorant_bilgi["Puan"].fillna(puan_mean)
print(new_df_restorant_bilgi["Puan"].isnull().sum())
new_df_restorant_bilgi=new_df_restorant_bilgi.to_csv("restorant_bilgileri.csv",index=False,encoding="utf-8-sig")
```

```python
mapping = {
    "Uygun fiyat aralığı": 1,
    "Orta fiyat aralığı": 2,
    "Yüksek fiyat aralığı": 3
}
new_df_restorant_bilgi["fiyat aralıgı kod"]=(
    new_df_restorant_bilgi["Fiyat Aralığı"].map(mapping)
)
```

```python
import re

def normalize_and_average(x):
    if pd.isna(x):
        return None
    
    # tüm tire varyasyonlarını "-" yap
    x = re.sub(r'[–—−]', '-', str(x))
    
    # metinden sadece rakamları al
    numbers = re.findall(r'\d+', x)
    
    if len(numbers) == 2:
        return (int(numbers[0]) + int(numbers[1])) / 2
    elif len(numbers) == 1:
        return int(numbers[0])
    else:
        return None

new_df_restorant_bilgi['Teslimat Süresi Dk'] = new_df_restorant_bilgi['Teslimat Süresi'].apply(normalize_and_average)

```

```python
new_df_restorant_bilgi=new_df_restorant_bilgi.drop(columns=["Teslimat Süresi"])
new_df_restorant_bilgi=new_df_restorant_bilgi.to_csv("restorant_bilgileri.csv",index=False,encoding="utf-8-sig")
```
